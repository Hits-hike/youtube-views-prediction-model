## Importing necessities

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

import pickle
```

## Reading data

```python
df = pd.read_csv('../data/face_dectection_data.csv')
df
```

## Preprocessing

```python
# 데이터 수집한 10000개에 대해서만
df = df.head(10000)
df
```

```python
# 결측값 있는 행 제거
df = df.dropna(subset=['people_num'])
df
```

```python
## Drop unnecessary columns
df = df.drop(columns = ['video_id', 'thumbnail'])
df
```

```python
# emotion과 gender을 원핫인코딩

df = pd.merge(df, pd.get_dummies(df['emotion'], dummy_na=True), left_index=True, right_index=True, how='left')
df = pd.merge(df, pd.get_dummies(df['gender'], dummy_na=True), left_index=True, right_index=True, how='left')

df = df.drop(columns = ['emotion', 'gender'])

df
```

## Train-Test Splitting

```python
X_data = df.drop(columns = ['views'])
y_data = df['views']

train_X, test_X, train_y, test_y = train_test_split(X_data, y_data, train_size = 0.80)
```

## Modeling

```python
model = RandomForestRegressor()
model.fit(train_X, train_y)

# validation
print(model.score(train_X, train_y))
print(model.score(test_X, test_y))
```
